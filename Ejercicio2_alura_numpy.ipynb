## Importaci√≥n del csv

```python
import numpy as np
path = '/content/citrus.csv'

datos = np.loadtxt(path, delimiter=',', usecols = np.arange(1,6,1), skiprows=1)
```

```python
diametros_naranjas = datos[1:5000,0]
diametros_toronjas = datos[5000:,0]

pesos_naranjas = datos[1:5000,1]
pesos_toronjas = datos[5000:,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametros_naranjas, pesos_naranjas)
plt.plot(diametros_toronjas, pesos_toronjas)
```
